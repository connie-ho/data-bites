```python
import pandas as pd
```

```python
import os
path = '/home/andy/Documents/data-bites/yelp_dataset_csv'
filepaths = []
for root, dirs, files in os.walk(path):
    for file in files:
        filepaths.append(os.path.join(root,file))
filepaths
```

```python
for csv in filepaths:
    print(csv.split('/')[-1], '\n', )
    for col in pd.read_csv(csv, nrows=1).columns:
        print(col)
```

```python
for csv in filepaths:
    print('\n', 'FILE_NAME:', csv.split('/')[-1] )
    df = pd.read_csv(csv, nrows=100000)
    for col in df.columns:
        print(df[col].value_counts())
```
